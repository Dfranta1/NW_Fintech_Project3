```python
# Initial imports
import pandas as pd 
import yfinance as yf
from datetime import datetime, timedelta
```

```python
# Take S&P500 tickers from Wikipedia 
payload = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
first_table = payload[0]
second_table = payload[1]

sp500 = first_table

sp500.head()
```

```python
# Separate tickers 
tickers = sp500['Symbol'].tolist()
```

```python
# Set timeframe 
current_date = datetime.now()
past_date = (datetime.now() - timedelta(90))
```

```python
# Download from Yfinance 
sp500_df = yf.download(tickers, past_date, current_date)
```

```python
# Visualize 
sp500_df.head()

# Make prices and returns dataframe 
prices = sp500_df.loc[:,'Adj Close']


returns = prices.pct_change()
returns = prices.iloc[1: , :]
returns.head()
```

```python
# Make Volume Dataframe 
volume = sp500_df.loc[:,'Volume']
volume = volume.iloc[1: , :]
```
